```python
# Try default UTF-8 first
destination = 'downloaded_file.csv'
import pandas as pd
try:
    df = pd.read_csv(destination)
except UnicodeDecodeError:
    # If encoding fails, try a more lenient encoding
    df = pd.read_csv(destination, encoding="latin1", header = None)

print(df.head())
```

```python
df[1:3]
```

```python
df = df.iloc[4:].reset_index(drop=True)  # remove first 4 rows and reset index
print(df.head())
```

```python
df.columns = df.iloc[0]   # Take first row as header
df = df[1:].reset_index(drop=True)  # Remove the first row from data

print(df.head())
```

```python
df.to_csv("processed_file.csv", index=False)
```
